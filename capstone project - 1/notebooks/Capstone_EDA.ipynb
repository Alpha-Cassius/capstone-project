## 1. Data Acquisition & Inspection

```
from data_cleaning import load_data

df = load_data('../data/raw/raw_client_data.csv')
display(df.head())
display(df.dtypes)
```

## 2. Null Value Analysis

```
from data_cleaning import analyze_nulls, fill_low_null_numeric_columns

null_df = analyze_nulls(df)
display(null_df)

# Fill numeric columns below 20% nulls with median
df = fill_low_null_numeric_columns(df, null_df)
```

## 3. Duplicate & Data Type Corrections

```
from data_cleaning import remove_duplicates, correct_dtypes

df = remove_duplicates(df)
df = correct_dtypes(df)
display(df.dtypes)
```

## 4. Descriptive Statistics & Skewness

```
from eda_analysis import get_descriptive_stats, calculate_skewness

display(get_descriptive_stats(df))
skewness, highest_skew = calculate_skewness(df)
display(skewness)
```

## 5. Visualizations

```
from visualizations import plot_line_chart, plot_bar_chart, plot_histogram, plot_scatter, plot_boxplot

plot_line_chart(df, 'Age')
plt.show()

plot_bar_chart(df, 'Pclass', 'Fare')
plt.show()

plot_histogram(df, highest_skew)
plt.show()

plot_scatter(df, 'Age', 'Fare')
plt.show()

plot_boxplot(df, 'Survived', 'Age')
plt.show()
```

## 6. Correlation Analysis

```
from eda_analysis import compute_correlations
from visualizations import plot_correlation_heatmap

pearson_corr, spearman_corr = compute_correlations(df)
plot_correlation_heatmap(pearson_corr)
plt.show()
```
